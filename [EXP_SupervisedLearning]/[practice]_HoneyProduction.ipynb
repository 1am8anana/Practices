# Honey Production

Now that you have learned how linear regression works, let’s try it on an example of real-world data.

As you may have already heard, the honeybees are in a precarious state right now. You may have seen articles about the decline of the honeybee population for various reasons. You want to investigate this decline and how the trends of the past predict the future for the honeybees.

Note: All the tasks can be completed using Pandas or NumPy. Pick whichever one you prefer.

```python
# import lib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

# read csv file(dataset)
df = pd.read_csv("honeyproduction.csv")

# show some head of dataset
print(df.head())

# get mean of total production per year
prod_per_year = df.groupby('year').totalprod.mean().reset_index()

# get column year in prod_per_year dataframe and reshape it to the right format
X = prod_per_year['year']
X = X.values.reshape(-1, 1)

# get dolumn totalprod in prod_per_year dataframe
y = prod_per_year['totalprod']

# plot it with scatter to see some relationship
plt.scatter(X, y)
plt.show()
```

### Create and Fit a Linear Regression Model

```python
# instantiate linear model using sklearn
model = linear_model.LinearRegression()

# fit features to the model
model.fit(X, y)

# show the slope and intercept of the line
print(model.coef_, model.intercept_)

# get the predictions
y_predict = model.predict(X)

# plot X and y_predict
plt.scatter(X, y)
plt.plot(X, y_predict)
plt.show()
```

### Predict the Honey Decline


So, it looks like the production of honey has been in decline, according to this linear model. Let’s predict what the year 2050 may look like in terms of honey production.

Our known dataset stops at the year 2013, so let’s create a NumPy array called X_future that is the range from 2013 to 2050.

```python
# create array that store year 2013 to 2050 using numpy array
X_future = np.array(range(2013, 2050))
# reshape it for sklearn
X_future = X_future.reshape(-1, 1)
```

* You can think of reshape() as rotating this array. Rather than one big row of numbers, X_future is now a big column of numbers — there’s one number in each row. (like a matrix transpose).

```python
print(X_future)
```

```python
# create a list of the predictions from this model
future_predict = model.predict(X_future)

# eplot the predictions of the future
plt.scatter(X, y)
plt.plot(X_future, future_predict, color = "red")
plt.show()
```

so this plot can be answering the question of How much honey will be produced in the year 2050, according to this?
